```python
from jiwer import cer
from jiwer import wer
def cer_wer(generated_text, ground_truth):
  cer_error = cer(ground_truth,generated_text)
  wer_error = wer(ground_truth,generated_text)
  return cer_error, wer_error
```

```python
def process_output(text):
  processed_out = extract_quoted_text(text)
  # print(processed_out)
  # print(len(processed_out.split(":")))

  if len(processed_out.split(":")) > 1:
    processed_out = processed_out.split(":")[1]


  if len(processed_out.split("is")) > 1:
    processed_out = processed_out.split("is")[1]
  return processed_out
```
